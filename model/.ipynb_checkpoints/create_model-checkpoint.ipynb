```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
#reading dataset
dataset = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
#preprocessing dataset
#RAM
dataset['Ram'] = dataset['Ram'].str.replace('GB','').astype('int32')
```

```python
#Weight
dataset['Weight'] = dataset['Weight'].str.replace('kg','').astype('float32')
```

```python
dataset.head(10)
dataset['Storage'].value_counts()
```

```python
#Storage
def add_storage(inpt):
    if inpt == '128 GB NVMe PCIe 3.0 x4 SSD' or inpt == '128 GB' or inpt == 'M.2 Slot for SSD Upgrade' or inpt == '128 GB SSD for Reduced Boot Up Time and in Game Loading':
        return '128 GB SSD'
    elif inpt == '1 TB HDD|512 GB SSD' or inpt == '512 GB SSD for Reduced Boot Up Time and in Game Loading' or inpt == '512 GB HDD|512 GB SSD':
        return '512 GB SSD'
    else:
        return inpt

dataset['Storage'] = dataset['Storage'].apply(add_storage)
    
dataset['Storage'].value_counts()
```

```python
#company
dataset['Company'].value_counts()
```

```python
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Razer' or inpt == 'Mediacom' or inpt == 'Microsoft' or inpt == 'Xiaomi' or inpt == 'Vero' or inpt == 'Chuwi' or inpt == 'Google' or inpt == 'Fujitsu' or inpt == 'LG' or inpt == 'Huawei':
    	return 'Other'
    else:
        return inpt

dataset['Company'] = dataset['Company'].apply(add_company)
    
dataset['Company'].value_counts()
```

```python
len(dataset['Product'].value_counts())
```

```python
dataset['TypeName'].value_counts()
```

```python
#ScreenResolution
dataset['ScreenResolution'].value_counts()
```

```python
dataset['Touchscreen'] = dataset['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
dataset['IPS'] = dataset['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
dataset.head()
```

```python
def add_screen_resolution(inpt):
    if inpt == 'Full HD 1920x1080' or inpt == 'IPS Panel Full HD 1920x1080' or inpt == 'IPS Panel Full HD / Touchscreen 1920x1080':
    	return '1920x1080 HD'
    elif inpt == '1366x768' or inpt == 'Touchscreen 1366x768' or inpt == 'IPS Panel 1366x768':
        return '1366x768'
    elif inpt == 'Quad HD+ / Touchscreen 3200x1800' or inpt == 'IPS Panel Quad HD+ / Touchscreen 3200x1800':
        return '3200x1800 HD+'
    elif inpt == 'IPS Panel 4K Ultra HD 3840x2160' or inpt == '4K Ultra HD 3840x2160':
        return '3840x2160 4K'
    else:
        return 'Other'

dataset['ScreenResolution'] = dataset['ScreenResolution'].apply(add_screen_resolution)
```

```python
dataset.head()
```

```python
#CPU
dataset['Cpu'].value_counts()
```

```python
len(dataset['Cpu'].value_counts())
```

```python
dataset['Cpu'].value_counts().head(30)
```

```python
dataset['CPUName'] = dataset['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
dataset['CPUName'].value_counts()
```

```python
dataset.head()
```

```python
def add_cpu_name(inpt):
    if inpt == 'Intel Core i7' or inpt == 'Intel Core i5' or inpt == 'Intel Core i3':
    	return inpt
    elif inpt.split()[2] == 'Celeron' or inpt.split()[2] == 'Pentium' or inpt.split()[2] == 'Core'  or inpt.split()[2] == 'Atom':
        return 'Intel'
    else:
        if inpt.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

dataset['CPUName'] = dataset['CPUName'].apply(add_cpu_name)
```

```python
dataset['CPUName'].value_counts()
```

```python
dataset['ClockSpeed'] = dataset['Cpu'].apply(lambda x:" ".join(x.split()[4:5]))
```

```python
dataset['ClockSpeed'].value_counts().head(100)
```

```python
dataset['ClockSpeed'] = dataset['ClockSpeed'].apply(lambda x:x if 'GHz' in x else '0GHz')
```

```python
dataset['ClockSpeed'].value_counts()
```

```python
dataset['ClockSpeed'] = dataset['ClockSpeed'].str.replace('GHz','').astype('float32')
```

```python
dataset['ClockSpeed'].value_counts()
```

```python
def add_clock_speed(inpt):
    if inpt >= 3.0:
    	return '3.00GHz<'
    elif inpt >= 2.0:
        return '2.00GHz<'
    elif inpt >= 1.0:
        return '1.00GHz<'
    else:
        return '1.00GHz>'

dataset['ClockSpeed'] = dataset['ClockSpeed'].apply(add_clock_speed)
```

```python
dataset['ClockSpeed'].value_counts()
```

```python
#gpu
dataset.head(3)
```

```python
dataset['Gpu'].value_counts().head(40)
```

```python
dataset['GPUName'] = dataset['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
dataset['GPUName'].value_counts()
```

```python
dataset = dataset[dataset['GPUName'] != 'ARM']
```

```python
dataset.shape
```

```python
#os
dataset.head(3)
```

```python
dataset['OpSys'].value_counts()
```

```python
def add_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
    	return 'Windows'
    elif inpt == 'Linux':
        return 'Linux'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    else:
        return 'Other'

dataset['OperatingSystem'] = dataset['OpSys'].apply(add_os)
```

```python
dataset['OperatingSystem'].value_counts()
```

```python
#Warranty
dataset['Warranty'].value_counts()
dataset['Warranty'] = dataset['Warranty'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
dataset['Warranty'].value_counts()
```

```python
def add_warranty(inpt):
    if inpt == '1' or inpt == 'One-year' or inpt == '24':
    	return '1 Year'
    elif inpt == '2':
        return '2 Years'
    elif inpt == '3':
        return '3 Years'
    elif inpt == '18':
        return '1.6 Years'
    else:
        return '1 Year'

dataset['Warranty'] = dataset['Warranty'].apply(add_warranty)
```

```python
dataset['Warranty'].value_counts()
```

```python
dataset['Warranty'].value_counts()
dataset.head(3)
```

```python
#dropping unneccessary cols
dataset = dataset.drop(columns=['laptop_ID', 'Product', 'Inches', 'Cpu', 'Gpu', 'OpSys'])
```

```python
#one hot encoding
dataset = pd.get_dummies(dataset)
```
